```python
import konlpy
from konlpy.tag import Okt
tagger = Okt()
```

```python
from gensim.models import Word2Vec
```

```python
import konlpy
txt = konlpy.utils.read_txt('./data/news.txt') 
txt
```

```python
import string
print(string.punctuation)
```

```python
txt = txt.translate(str.maketrans('', '', string.punctuation))
txt
```

```python
txt = txt.replace('\n', '')
txt
```

```python
noun = tagger.nouns(txt)
noun2more = [ele for ele in noun if len(ele) > 1]
noun2more
```

```python
# model = Word2Vec(data,         # 리스트 형태의 데이터
#                  sg=1,         # 0: CBOW, 1: Skip-gram
#                  size=100,     # 벡터 크기
#                  window=3,     # 고려할 앞뒤 폭(앞뒤 3단어)
#                  min_count=3)  # 사용할 단어의 최소 빈도(3회 이하 단어 무시)


model = Word2Vec([noun2more], size=300, window=5, min_count=3, sg=1)
```

```python
result = model.most_similar('게임')
result
```

```python
cos = model.wv.similarity('게임', '질병')
print(cos)
```
